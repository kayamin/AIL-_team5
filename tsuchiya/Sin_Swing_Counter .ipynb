## 振幅回数判別モデル
LSTMを使ってできるか実験

```python
from keras.models import Sequential
from keras.layers.core import Dense, Activation, Dropout, Flatten
from keras.layers.recurrent import LSTM
from keras.optimizers import Adam, RMSprop
from keras.callbacks import EarlyStopping
from keras.layers.convolutional import Conv1D, MaxPooling1D
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
# Set
k_num = [0,1,2,3] # 筋トレ回数
T = 100 # フレーム数
N = 5000
use_re = True
```

```python
def make_dataset(k_num, N=2000, T=100, use_re=False, ampl=0.05):
    '''
    一定のフレーム数ないでの筋トレ回数をランダムで設定し， 
    ノイズを加えて，trainデータを生成する
    '''
    data = []
    target = np.random.choice(k_num, N, p = [0.15, 0.6, 0.2, 0.05])
    
    for n in target:
        n_noise = (np.random.randint(0,10) - 5) / 40.0
        cos_noise = 0.5 + ampl * np.random.uniform(low=-1.0, high=1.0, size=100)
        
        if n !=0:
            if not use_re:
                x = np.arange(0, T)
                x =  np.sin( ((n+n_noise) * 2.0* np.pi * x) / len(x) ) * cos_noise
            else:
                redu = np.random.randint(0, 20)
                forb = np.random.choice([True, False])
                x = np.arange(0, T-redu)
                temp = np.ones(redu)
                x =  np.sin( ((n+n_noise) * 2.0* np.pi * x) / len(x)  - np.pi / 2 )
                if forb:
                    x = np.append(x, temp*x[-1]) * cos_noise
                else:
                    x = np.append(temp*x[0], x) * cos_noise

        else:
            x = np.arange(0, T)
            x = np.cos( 0 * np.pi * x / len(x) )  * cos_noise
        
        x -= x.min()
        x /= x.max()
        data.append(x)

    re_data = np.array(data).reshape(N, T, 1)
    re_target = target.reshape(N, 1)
    return re_data, re_target

#g -> 学習データ，h -> 学習ラベル
g, h = make_dataset(k_num, N, T, use_re, 0.05)
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder()
h = enc.fit_transform(h).toarray()
```

```python
batch_size=300
epochs=100
validation_split=0.1
lr = 0.001

model = Sequential()
model.add(Dense(256, activation='relu', input_shape=(100,)))
model.add(Dropout(0.2))
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(len(k_num), activation='softmax'))

model.summary()
model.compile(loss='categorical_crossentropy',
              optimizer=RMSprop(),
              metrics=['accuracy'])
```

```python
batch_size=300
epochs=100
validation_split=0.1
lr = 0.001

model = Sequential()
model.add(Conv1D(256, 10, activation='relu', input_shape=(100, 1)))
model.add(MaxPooling1D(pool_size=2))
model.add(Flatten())
model.add(Dropout(0.2))
model.add(Dense(len(k_num), activation='softmax'))

model.summary()
model.compile(loss='categorical_crossentropy',
              optimizer=RMSprop(),
              metrics=['accuracy'])
```

```python
# Go !
early_stopping = EarlyStopping(monitor='val_loss', mode='auto', patience=20)
model.fit(g, h,
          batch_size=batch_size,
          epochs=epochs,
          validation_split=validation_split,
          callbacks=[early_stopping]
          )
```

```python
# 予測
y_data, y_label = make_dataset(k_num, 1000, T, use_re, 0.05)
predicted = model.predict(y_data).astype(int).argmax(axis=1)
```

```python
print(confusion_matrix(y_label.flatten(), predicted.flatten()))
acc = accuracy_score(y_label.flatten(), predicted.flatten())* 1000
print(int(acc)/10.0, '%')
```
