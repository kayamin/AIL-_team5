```python
import numpy as np
import pandas as pd
import glob
```

## OpenPoseが動かなかったからか画像名しか入っていない行を、全ての値が0にして列数が他の行と揃うようにする

```python
directory = "../data/movie_push-ups-from-side/2-1.COCO-pose/"
save_dir = "../data/movie_push-ups-from-side/2-1.COCO-pose_mod/"
files = glob.glob(directory + "*.csv")
```

```python
f = files[1]
print(f)
df = pd.read_csv(f, header=None, dtype=str)
df
```

```python
idx = np.arange(55)
idx[0], idx[-1] = 54, 0
print(idx)
df[df.isnull().any(axis=1)] = df[df.isnull().any(axis=1)].iloc[:, idx]
df = df.replace(np.nan, "0")
```

```python
idx = np.arange(55)
idx[0], idx[-1] = 54, 0
for f in files:
    df = pd.read_csv(f, header=None, dtype=str)
    df[df.isnull().any(axis=1)] = df[df.isnull().any(axis=1)].iloc[:, idx]
    df = df.replace(np.nan, "0")
    save_f = f.replace(directory, save_dir)
    df.to_csv(save_f, header=False, index=False)
```
